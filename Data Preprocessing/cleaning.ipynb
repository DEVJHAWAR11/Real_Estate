```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if isinstance(x, float):
        return x
    if isinstance(x, list) and len(x) >= 2:
        if x[1] == 'Lac':
            return round(float(x[0]) / 100, 2)
        else:
            return round(float(x[0]), 2)
    return None  # fallback for invalid/malformed input
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.sample(5)
```

```python
df["price-per-sqft"]
```

```python
df['price-per-sqft'] = df['price-per-sqft'].str.split("/").str.get(0).str.replace("₹"," ").str.replace(",","").str.strip()
```

```python
df=df[~df['bedRoom'].isnull()]
```

```python
df['bedRoom'] = df['bedRoom'].str.split(" ").str.get(0).str.strip()
```

```python
df = df[~df['bathroom'].isnull()]
```

```python
df['bathroom'] = df['bathroom'].str.split().str.get(0).str.strip()
```

```python
df = df[~df['balcony'].isnull()]
```

```python
df['balcony'].str.split().str.get(0).str.replace('No','0')
```

```python
df['balcony'] = df['balcony'].str.split().str.get(0).str.strip()
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['floorNum'].str.split(" ").str.get(0).value_counts()
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df = df[~df['price-per-sqft'].isnull()]
```

```python
df['price-per-sqft']
```

```python
# Ensure you're working on the original DataFrame, not a slice
df.loc[:, 'price-per-sqft'] = pd.to_numeric(
    df['price-per-sqft'].str.replace(',', '', regex=False), 
    errors='coerce'
)
```

```python
df.insert(loc=4,column='area',value=round((df['price']*10000000)/df['price-per-sqft']))
```
